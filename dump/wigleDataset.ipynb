```python
import graphlab
from pandas import read_csv
import pandas as pd
from os import listdir, chdir, path
import json
import csv
```

```python
chdir('/Users/nmohan/Documents/Git Code/Wigle dataset/dump')

dataset_name = 'combined_dataset.csv'
```

```python
file_index = 0
for fileindex in range (0, 42):
    filepath = "wiggle" + str(fileindex) + ".json"
    
    firsttime = True
    with open(filepath) as json_data:
        d = json.load(json_data)             #load all data from json

        for key, value in d.items():
            if(key=="results"):
                dataset = value             #get only results section from data in the datset
                
                keys = dataset[0].keys()
                alldata = pd.DataFrame(dataset, columns=keys)
                
                with open(dataset_name, 'a') as f:
                    if(firsttime == True):
                        alldata.to_csv(f, header=True, index=False, encoding='utf-8')
                        firsttime = False
                    else:
                        alldata.to_csv(f, header=False, index=False, encoding='utf-8')
                    
                fileindex += 1
```
